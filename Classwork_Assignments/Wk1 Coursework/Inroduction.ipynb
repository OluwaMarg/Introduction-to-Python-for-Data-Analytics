### C. Hypothesis Testing

##### 7a. Statistically significant difference in sepal length among the different species:
#####    -One-Way ANOVA (Assuming data is Normal and Variances are homogenous- ie. equal)

```python
# Prepare data
x = iris[['SepalLengthCm']] # feature
y = iris['Species'] # target
# Run one-way ANOVA
f_stat, p_value = f_classif(x, y)
# Print result
print(f'ANOVA F-statistics: {f_stat[0]:.4f}, \nP-value: {p_value[0]:.4f}')
# Interpretation
if p_value < 0.05:
    print("Result: there is a statistically significant difference in sepal length among the different species (reject the Ho)")
else:
    print("Result: No statistically significant difference in sepal length among the different species (fail to reject the Ho)")
```

##### 7b. Visualization: Boxplot for Spread and difference in Sepal length among the different Species 

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.title('Distribution of Sepal Length Across Species')
plt.show()
```

##### 8a. t-test to compare the means of two species for petal width

```python
# Select the two Species eg Iris-Setosa and Iris-versicolor
# Filter the DataFrame to include only these two Species

iris_Iris_Setosa = iris[iris['Species'] == 'Iris-setosa']
iris_Iris_Versicolor = iris[iris['Species'] == 'Iris-versicolor']
# Perform t-test
t_test, p_value = ttest_ind(iris_Iris_Setosa['PetalWidthCm'], iris_Iris_Versicolor['PetalWidthCm'])
print(f'T-Statistics: {t_test}, \nP-Value: {p_value}')
# Significance Level (alpha)
alpha = 0.05
# Interpretation
if p_value < alpha:
    print("Result: there is a statistically significant difference in Petal Width between Iris-setosa and Iris-versicolor Species (reject the Ho)")
else:
    print("Result: No statistically significant difference in Petal Width between Iris-setosa and Iris-versicolor Species (fail to reject the Ho)")
```

##### 10. Decision Boundaries Plot- for the classification of the species using a pair of features

```python
f = iris[['SepalLengthCm', 'SepalWidthCm']] # choosing pair of features
t = iris['Species']        # target
t_encoded = LabelEncoder().fit_transform(t)  # encoded label into numbers

model = SVC(kernel='linear') # Train a classifier (SVM with linear kernel)
model.fit(f, t_encoded)

a_min, a_max = iris['SepalLengthCm'].min() - 0.5, iris['SepalLengthCm'].max() + 0.5
b_min, b_max = iris['SepalWidthCm'].min() - 0.5, iris['SepalWidthCm'].max() + 0.5
aa, bb = np.meshgrid(np.linspace(a_min, a_max, 200), np.linspace(b_min, b_max, 200)) # Create a mesh grid using Numpy
mesh_df = pd.DataFrame(np.c_[aa.ravel(), bb.ravel()], columns=['SepalLengthCm', 'SepalWidthCm']) # Convert mesh grid to Pandas DataFrame for prediction

Z = model.predict(mesh_df)  # Predict class for each point in the mesh grid using trained model
Z = Z.reshape(aa.shape) 

plt.figure(figsize=(8, 6))
plt.contourf(aa, bb, Z, alpha=0.3, cmap='coolwarm') # Plot decision boundaries (decision regions) using contourf() to shade the regions 

sns.scatterplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], hue=iris['Species'], palette='coolwarm', edgecolor='k') # Scatter plot for better visualize

plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title(f'Decision Boundaries for Sepal Length and Sepal Width')
plt.legend(title='Species')
plt.show()
```

##### 11b. Handle Missing values

```python
iris_cleaned_rows = iris.dropna() # If missing values are few, we can simply drop those rows
print(iris_cleaned_rows)
iris_cleaned_columns = iris.dropna(axis=1) # If an entire column has many missing values, you can drop it
print(iris_cleaned_columns)
iris_filled_num_col = iris.fillna(iris.iloc[:, [1, 2, 3, 4]].mean()) # If missing values are in important numerical columns
print(iris_filled_num_col)
iris_filled_cat_col = iris.fillna(iris['Species'].mode()) # If missing values are in categorical columns, fill them with the most frequent value
print(iris_filled_cat_col)
```

##### 12a. Identifying Outliers: Interquantile Range (IQR) Method (Statistical approach)

```python
Q1 = iris.iloc[:, [1, 2, 3, 4]].quantile(0.25)
Q3 = iris.iloc[:, [1, 2, 3, 4]].quantile(0.75)
IQR =Q3 - Q1
iris_outliers = ((iris.iloc[:, [1, 2, 3, 4]] < (Q1 - 1.5 * IQR) | (iris.iloc[:, [1, 2, 3, 4]] > (Q3 + 1.5 * IQR)))) # Result display as Tuple of bool
print(iris_outliers.sum()) # Count of outliers in each column
# print(iris_outliers.count()) # Incorrect
print(iris_outliers)
```

##### 12a. Identifying Outliers: Box Plot Method (Visualization approach)

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris.iloc[:, [1, 2, 3, 4]]) 
plt.title("Boxplot of Iris Features to Detect Outliers")
plt.show()            # points outside the whiskers are potential outliers
```

##### 12b. Managing Outliers- Method 1: Remove Outliers

```python
iris_no_outliers1 = iris.iloc[:, [1, 2, 3, 4]][ ~ ((iris.iloc[:, [1, 2, 3, 4]] < (Q1 - 1.5 * IQR) | (iris.iloc[:, [1, 2, 3, 4]] > (Q3 + 1.5 * IQR))))]
iris_no_outliers2 = iris.iloc[:, [1, 2, 3, 4]][ ~ iris_outliers]      # same as line above
print(iris_no_outliers1)
print(iris_no_outliers2) # Same as line above
print(iris.merge(iris_no_outliers1)) # what does the merge() method does? 48 rows seems to be missing after merge.
```

##### 12b. Managing Outliers- Method 2: Replace Outlier with Median

```python
iris_no_outliers = iris.iloc[:, 1:5].copy()    # Create a copy of the original DataFrame to be modified

for col in iris_no_outliers:
    median_value = iris_no_outliers[col].median()
    iris_no_outliers[col] = iris_no_outliers[col].mask(((iris_no_outliers < (Q1 - 1.5 * IQR) | (iris_no_outliers > (Q3 + 1.5 * IQR))))[col], median_value)
   # iris_no_outliers[col] = iris_no_outliers[col].mask(iris_outliers[col], median_value) # Same as above line
print(iris_no_outliers.head())  # Outliers replaced with median
```

##### 15b. Prepare iris dataset for machine learning

```python
# Check for missing values
iris_isnull = iris.isnull().sum()
print(iris_isnull)

# Encode categorical variables

le = LabelEncoder()                 # Creates an instance of LabelEncoder
iris['species_encoded'] = le.fit_transform(iris['Species'])

# Split the dataset into features and encoded target
f = iris.drop(iris[['Id', 'Species', 'species_encoded']], axis=1)      # features 
t = iris['species_encoded']      # encoded target

# Split the dataset into training (80%) and testing (20%) sets
f_train, f_test, t_train, t_test = train_test_split(f, t, test_size=0.2, random_state=42)       # using train_test_split of sklearn

# Scale the features using StandardScaler                # This preprocessing step was NOT done in splitting dataset to training & testing sets above

scaler = StandardScaler()                     # Creates an instance of StandardScaler
f_train = scaler.fit_transform(f_train)         # Fit on training data, then transform
f_test = scaler.transform(f_test)               # Transform test data using the same scaler
# Now your iris dataset is ready for machine learning modeling

print()
print(f_train) 
print()
print(f_test)
print()
print(t_train)
print()
print(t_test)
```

##### 16. Analysis: feature importance; Model: Random Forest for classification

```python
# Continuation of code snippets on preparing iris dataset for machine learning

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)   # Initialize model (create instance)
rf_model.fit(f_train, t_train)                                         # Train model
feature_importance = rf_model.feature_importances_               # extract feature importance from model's attribute (store)
print(feature_importance)
iris_importance = pd.DataFrame({'Feature': f.columns, 'Importance': feature_importance})    # Create DataFrame for better visualization
iris_importance = iris_importance.sort_values(by='Importance', ascending=False)
print()
print(iris_importance)
plt.figure(figsize=(8, 5))                                        # Visualize feature importance on Horizontal Bar Plot
plt.barh(iris_importance['Feature'], iris_importance['Importance'], color='skyblue')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importance in Random Forest Model')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature on top
plt.show()
```

### F. Time Series Analysis

##### 17a. Generate date column in iris dataset

```python
num_rows = iris.shape[0]             # Define the number of rows in the dataset.  Outputs the first value in the tuple generated
print(num_rows)

date_range = pd.date_range(start='2024-01-01', end='2024-08-31', periods=num_rows)    # Generate random date. We spread the date within 8 months.
print()
print(date_range)
np.random.shuffle(date_range.values)         # Randomizes and shuffles the dates to simulate real-world unordered data.
iris['Date'] = date_range
# print(date_range)
iris['Date'] = pd.to_datetime(iris['Date'])          # Convert time column to datetime format
iris.set_index('Date', inplace=True)                  # Set time column as index
```

##### 17b. Resampling for Trends/Time-based Grouping (eg. Aggregating by month)

```python
monthly_trend_mean = iris.drop(columns=['Id', 'Species', 'species_encoded']).resample('ME').mean()   
print()
print(monthly_trend_mean)
```

##### 17c. Rolling (Moving) Averages for Smoothing Trends

```python
iris = iris.sort_values('Date')                       # Sort iris DataFrame in Ascending order (default)
iris['SepalLengthCm_MA'] = iris['SepalLengthCm'].rolling(window=7).mean()   # 7-days Rolling Average
print()
print(iris['SepalLengthCm_MA'])
```

##### 17d. Visualizing Time Series Trends

```python
plt.figure(figsize=(10, 5))
plt.plot(iris.index, iris['SepalLengthCm'], label='Sepal Length')
plt.plot(iris.index, iris['SepalLengthCm_MA'], label='7-day MA', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Sepal Length (cm)')
plt.title('Trend Over Time')
plt.legend()
plt.show()
```

##### 17e. First-Order Differencing (Remove Linear Trend) Useful for making data suitable for forecasting models like ARIMA

```python
iris['SepalLengthCm_Diff'] = iris['SepalLengthCm'].diff()  # First difference

# Plot original and differenced series
plt.figure(figsize=(10, 5))
plt.plot(iris['SepalLengthCm'], label='Original')
plt.plot(iris['SepalLengthCm_Diff'], label='Differenced', linestyle='dashed')
plt.legend()
plt.show()
```

##### Decomposition

```python
# Decomposition: Helps identify underlying trends, seasonal effects, and random (irregular or residual) noise in the data.
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the Sepal Length time series
decomposition = seasonal_decompose(iris['SepalLengthCm'], model='additive', period=30)  

# Plot the decomposed components
decomposition.plot()
plt.show()
```
